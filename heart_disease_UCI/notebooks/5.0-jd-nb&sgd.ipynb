```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('heart_disease_cleaned.csv')
df.head()
```

```python
df.drop('Unnamed: 0',axis=1, inplace = True)
df.head()
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report,confusion_matrix,auc,roc_auc_score,roc_curve
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import multilabel_confusion_matrix
```

```python
from sklearn.metrics import balanced_accuracy_score
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
X = df.drop('Target',axis=1)
y = df['Target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40)
```

```python
scaler = StandardScaler()
scaler.fit(df.drop('Target',axis=1))
```

```python
scaled_features = scaler.transform(df.drop('Target',axis=1))
```

```python
df_feat = pd.DataFrame(scaled_features,columns=df.columns[:-1])
```

```python
df_feat.head()
```

```python
X = df_feat
y = df['Target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40)
```

```python
sgd_clf = SGDClassifier()
sgd_clf.fit(X_train,y_train)
```

```python
gnb = GaussianNB().fit(X_train, y_train)
gnb_predictions = gnb.predict(X_test)
  
cm = confusion_matrix(y_test, gnb_predictions)
```

```python
pred = sgd_clf.predict(X_test)

print('Stochastic Gradient Descent:') 
print('\n')
print(multilabel_confusion_matrix(y_test, pred))
print('\n')
print(classification_report(y_test,pred))
print(balanced_accuracy_score(y_test,pred))
print('\n')
print('\n')
print('Naive Bayes:')
print('\n')
print(multilabel_confusion_matrix(y_test, gnb_predictions))
print('\n')
print(classification_report(y_test,gnb_predictions))
print(balanced_accuracy_score(y_test,gnb_predictions))
```
